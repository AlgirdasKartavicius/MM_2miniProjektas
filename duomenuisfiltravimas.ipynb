```python
import pandas as pd
data = pd.read_csv('at20190929.csv', sep=',', error_bad_lines=False, index_col=False, encoding="ISO-8859-1") 



data['PIRM_REG_DATA_LT'] =pd.to_datetime(data['PIRM_REG_DATA_LT'], errors='coerce')

data3 = data['PIRM_REG_DATA_LT'] > '2016'


filtered =data[data3]
```

```python
filtered.head()
```

```python


data = filtered[['MARKE', 'PIRM_REG_DATA_LT', 'PASKUTINES_REG_DATA',
                 'DEGALAI', 'KILMES_SALIS', 'SANDORIO_SUMA', 'VALIUTA', 'SANDORIO_DATA',
                'GALIA', 'DARBINIS_TURIS', 'PIRM_REG_DATA']]
data
```

```python
filteredBIn = data .groupby([data ['PIRM_REG_DATA_LT'].dt.year,'MARKE'])['MARKE'].count().to_frame('Count').reset_index()
data3 = filteredBIn['Count'] > 2000
filtered4 =filteredBIn[data3]
markes = set()
for x in filtered4['MARKE']:
    markes.add(x)
markes
```

Buvo pastebėta, kad tarp populiariausių nėra *VOLKSWAGEN* ir *MERCEDES BENZ* automobilių, nors regitros puslapyje jie yra tarp populiariausių. Nustatyta, kad ne visos automobilių markės parašytos teisingai (identiški pavadinimai), todėl buvo atliktas duomenų valymas (pavadinima pakeisti taip, kad būtų vienodi tai automobilių markei)

```python
data['MARKE'].replace(to_replace='MERCEDES-BENZ', value='MERCEDES BENZ', inplace= True)
data['MARKE'].replace(to_replace='VOLKSWAGEN-VW', value='VOLKSWAGEN', inplace= True)
data['MARKE'].replace(to_replace='VW', value='VOLKSWAGEN', inplace= True)
data['MARKE'].replace(to_replace='VOLKSWAGEN. VW', value='VOLKSWAGEN', inplace= True)
```

```python

filteredBIn = data .groupby([data ['PIRM_REG_DATA_LT'].dt.year,'MARKE'])['MARKE'].count().to_frame('Count').reset_index()
data3 = filteredBIn['Count'] > 2000
filtered4 =filteredBIn[data3]
markes = set()
for x in filtered4['MARKE']:
    markes.add(x)
markes
```

```python
data.to_csv ('exported_car_model_data.csv', index = None, header=True,  sep=',', ) 
```
